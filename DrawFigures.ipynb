```python
%load_ext autoreload
%autoreload 2

%matplotlib inline

from __future__ import division

import matplotlib
import numpy as np
import matplotlib.pyplot as plt


from visiprog.data import get_old_visiprog_coverage, get_varied_visiprog_coverage
```

```python
matplotlib.rcParams.update({'font.size': 14})

fig = plt.gcf()
fig.set_size_inches(10, 3.5)

ax1 = plt.subplot(121)

c1 = get_old_visiprog_coverage()
c1 = c1[:225]
plt.plot(c1, label='Original ViSiProg')
plt.xlabel("Number of trials")
plt.ylabel("Percentage of coverage")
plt.legend(loc=2)
ax2 = plt.subplot(122, sharey=ax1, sharex=ax1)

c2 = get_varied_visiprog_coverage()
c2 = c2[:225]
plt.plot(c2, label="Modified ViSiProg")
plt.xlabel("Number of trials")
plt.legend(loc=2)
fig.suptitle("Comparison of convergence")


# plt.tight_layout()

fig.savefig("convergence.pdf", bbox_inches='tight')
```
