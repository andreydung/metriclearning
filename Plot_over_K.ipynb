```python
import pandas as pd

%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

from IPython.display import Image, display
```

```python
STSIMM = []
VSP_original = []
VSP = []
material = []

k_list = range(1,11)

for k in k_list:
    print(k)
#     filepath = "thrasos/within_fromcluster_k{}.csv".format(k)
    filepath = "thrasos/within_k{}.csv".format(k)
    # filepath = "thrasos/results/across_material_kmeans/across_trainedongroups_50material_k{}.csv".format(k)
    
    df = pd.read_csv(filepath, index_col=0)
    ave = df.loc['average']
   
    STSIMM.append(ave['STSIM-M'])
    VSP_original.append(ave['visiprog_original'])
    VSP.append(ave['visiprog'])
    material.append(ave['material'])
```

```python
matplotlib.rcParams.update({'font.size': 25})

fig = plt.gcf()
fig.set_size_inches(13, 8)


plt.plot(k_list, material, '-ro', label='semantic')
plt.plot(k_list, STSIMM, '-ko', label='STSIM-M')
plt.plot(k_list, VSP, '-o', label='Modified LFDA')
plt.plot(k_list, VSP_original, '-bo', label='LFDA')

plt.xticks(k_list)
plt.xlabel('Number of exemplars per cluster')
plt.ylabel('Precision@1')

plt.legend()

# plt.title("Trained from ViSiProG groups")

fig.savefig('thrasos/within.pdf')
fig.savefig('thrasos/within.eps', dpi=1200)
```

```python
df = pd.DataFrame({'k':k_list, 'STSIM-M':STSIMM,'visiprog':VSP, \
                   'visiprog_original': VSP_original, 'material':material}, \
                  columns=['k','STSIM-M','visiprog_original','visiprog','material'])
df = df.set_index('k')
display(df)

df.to_csv('thrasos/across.csv')
```
